Product-market fit:
Relation of value proposition and customer segment (from Business Model Canvas).
Someone needs something, our product satisfies that need - AKA pain or gain.
Human factors: 
1. Needs (task analysis, user mapping/flow). Quantifying in terms of time to completion, satisfaction.
2. Values (pens model -- non-goal oriented behavior, **connectedness**, autonomy, mastery). Book: Symbols for sale (selling people's identity back to them). Documentary: Objectified.
3. Capabilities: extend or amplify
4. Limitations: complement (attention/memory)

Accurate Empathy: the greatest trait a PO can have for the end-user.

Cognitive baggage: when past experience interferes or enhances present experience. Accurate empathy requires understanding users cognitive baggage: you must meet ones expectation before exceeding or transforming them.

Intuition: gut feeling based on past experience.

To build intuitive software, a PO must have accurate knowledge of similar systems and the uer's training or past experience in order to have accurate empathy.

Mapping: take something that user doesn't know and map it to something the user do know.
Nesting: strategy to minitage information overload.

What criteria do you consider when you apply a value judgement (good/bad) to a product, service or experience?

Usability: information overload, trainign(higher/low)
Brand fidelity and reputation (influence expectations, openness)
Performance: speed (moment to moment latency in milliseconds), reliability, consistency
Perceived efficiency (that could've been easier): number of screen or steps

The cardinal rule of human factors (in which we design the work to match the worker): capacity > demand.

Increase capacity (training, onboarding) or decrease (automated paths, limited functionality) - friendly for novice but often frustrating for expert users.

Usability-functionality tradeoff: the tension between users and systems of varying capability, generally, is called the usability-functionality tradeoff.

Objectives measures of performance: 
- Task analysis for goal-oriented interfaces:
    - Task time to completion
    - Error rate (errors of commission, errors of omission)
- Systems usability survey: SUS (system usability scale) survey.
- USE survey.
- Cognitive walkthrough
- interviews
- ethnographic (value driven)
