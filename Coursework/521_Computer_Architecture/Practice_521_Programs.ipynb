Less than 11

```python
# if less than or equal to 11
# py program

def le11py(inx):
    outy = 0
    if inx <= 11:
        outy = 1
    return outy

le11asm = """
#: inx = r1 # r1 for input
#: outy = r2 # r2 for output

Set(outy, 0) # outy = 0
# if inx <= 11: # (if inx > 11 goto endif)
Set(cr, 11)
Sub(cr, cr, inx)
Addi(ar, pc, ?endif) # ar = address of endif
Moven(pc, ar, cr) # if inx > 11, go to endif
    Set(outy, 1) # else, change outy to 1
#> endif
Move(outy, outy) #print(outy)
"""
```

Absolute value

```python
def abspy(r1)->r2:
    if r1 < 0:
        r1 = -r1
    return r1

absasm = """
#> absasm(r1)->r2:
    #: inx = r1
    #: outy = r2
    Move(outy, inx)
    Addi(ar, pc, ?endif)
    # if r1 < 0: if r1 >= 0 gotoendif
    Movep(pc, ar, inx)
        Not(outy, outy)
        Addi(outy, outy, 1)
    #>endif
    Move(r2, r2)
    # return r2
"""
```

```python
test = -10
print(abspy(test))

start(absasm)
setr(r1, test)
print(getr(r2))
```

Whileloop

```python
def whilepy(inx:r1)->r2:      # pass by reference :r1, :r2
    outy:r2 = 0                 # initialize output = 0
    while inx >= 0:
        outy = outy + 2
        inx = inx - 1           # decreases to 0
    return outy

whileasm = """
#>whileasm(inx:r1)->r2:      # pass by reference :r1, :r2
#: inx = r1
#: outy = r2
Set(outy, 0)
Addi(ar, pc, ?endwhile)
#>while inx >= 0: if inx <0 goto endwhile
Moven(pc, ar, inx)
    Addi(outy, outy, 2)
    Subi(inx, inx, 1)
    Subi(pc, pc, ?while)
#>endwhile
#return outy

"""
```

```python
test = 10
print(whilepy(test))

setr(r1, test)
start(whileasm)
print(getr(r2))
```

Positive multiply

```python
def mulpy(r1, r2)->r3:
    r3 = 0
    while r2!= 0:
        r3 += r1
        r2 -= 1
    return r3

mulasm = """
#> mulasm(r1, r2)->ans:
#: inx = r1
#: iny = r2
Set(ans, 0)
Addi(ar, pc, ?endwhile)
#>while r2!= 0: if r2 == 0 gotoendwhile
Movez(pc, ar, iny)
    Subi(iny, iny, 1)
    Add(ans, ans, inx)
    Subi(pc, pc, ?while)
#>endwhile
# return ans
"""
```

```python
testx = 2
testy = 0

print(mulpy(testx, testy))

setr(r1, testx)
setr(r2, testy)
start(mulasm)
print(getr(ans))
```

+/- Multiply

```python
def mulpy(r1, r2)->r3:
    r3 = 0
    if r2 >= 0:
        while r2!= 0:
            r3 += r1
            r2 -= 1
    else:
        while r2!= 0:
            r3 -= r1
            r2 += 1
    return r3

mulasm="""
#> mulasm(r1, r2)->ans :
#: ina = r1
#: inb = r2
Set(ans, 0)
Addi(ar, pc, ?else)
    #if r2 >= 0: if r2 < 0 goto else
        Moven(pc, ar, inb)
        Addi(ar, pc, ?endwhile)
        #>while1 r2!= 0: if r2 == 0 goto endwhile
            Movez(pc, ar, inb)
            Add(ans, ans, ina)
            Subi(inb, inb, 1)
            Subi(pc, pc, ?while1)
    #>else:
        Addi(ar, pc, ?endwhile)
        #>while2 r2!= 0: if r2 == 0 goto endwhile
            Movez(pc, ar, inb)
            Sub(ans, ans, ina)
            Addi(inb, inb, 1)
            Subi(pc, pc, ?while2)

#>endwhile
# return ans

"""
```

```python
testx = 2
testy = 3

setr(r1, testx)
setr(r2, testy)
startfast(mulasm)
py = mulpy(testx, testy)
asm = getr(ans)
if py == asm:
    print("working", py , asm)

testx = -2
testy = 3

setr(r1, testx)
setr(r2, testy)
startfast(mulasm)
py = mulpy(testx, testy)
asm = getr(ans)
if py == asm:
    print("working", py , asm)

testx = -2
testy = -3

setr(r1, testx)
setr(r2, testy)
startfast(mulasm)
py = mulpy(testx, testy)
asm = getr(ans)
if py == asm:
    print("working", py , asm)

testx = 2
testy = -3

setr(r1, testx)
setr(r2, testy)
startfast(mulasm)
py = mulpy(testx, testy)
asm = getr(ans)
if py == asm:
    print("working", py , asm)
```

Division

```python
def divpy(ina:r1, inb:r2)->(ans, res):
    ans = 0
    res = 0
    if inb != 0:
        res = ina
        while res >= inb:
            res -= inb
            ans += 1
    return ans, res

divasm = """
#> divasm(ina:r1, inb:r2)->(ans, res):
#: ina = r1
#: inb = r2
Set(ans, 0)
Set(res, 0)
# if inb != 0: if inb == 0, goto end
Addi(ar, pc, ?end)
Movez(pc, ar, inb)
    Move(res, ina)
    Addi(ar, pc, ?end)
    #>while res >= inb: if res < inb goto end
    Sub(cr, res, inb)
    Moven(pc, ar, cr)
        Sub(res, res, inb)
        Addi(ans, ans, 1)
        Subi(pc, pc, ?while)
#>end
"""
```

```python
testx = 4
testy = 0

setr(r1, testx)
setr(r2, testy)
startfast(divasm)
py = divpy(testx, testy)
asm = getr(ans), getr(res)
if py == asm:
    print("working", py , asm)

testx = 4
testy = 2

setr(r1, testx)
setr(r2, testy)
startfast(divasm)
py = divpy(testx, testy)
asm = getr(ans), getr(res)
if py == asm:
    print("working", py , asm)

testx = 4
testy = 4

setr(r1, testx)
setr(r2, testy)
startfast(divasm)
py = divpy(testx, testy)
asm = getr(ans), getr(res)
if py == asm:
    print("working", py , asm)

testx = 4
testy = 5

setr(r1, testx)
setr(r2, testy)
startfast(divasm)
py = divpy(testx, testy)
asm = getr(ans), getr(res)
if py == asm:
    print("working", py , asm)
else:
    print("not working", py, asm)
```

Fibo

```python
def fib_rec(n:r1)->ans:
	if n < 2:
		f:ans = 1
	else:
		f = fib(n-1) + fib(n-2)
	return f

def fib(n:r1)->ans:
	f:ans = 1
	last_1_sum:r2 = 1
	last_2_sum:r3 = 1
	counter:r4 = 1
	if n!=1:
		while counter < n:
			f = last_1_sum + last_2_sum
			last_2_sum = last_1_sum
			last_1_sum = f
			counter += 1
	return f
	
fibasm = """
#>fibasm(n:r1)->ans:
Set(ans, 1)
Set(r2, 1)
Set(r3, 1)
Set(r4, 1)

Subi(cr, r1, 1)
Addi(ar, pc, ?end)
#>if r1!=1: if r1 == 1 goto end
Movez(pc, ar, cr)
	Addi(ar, pc, ?end)
	#>while r4 < r1: if r4 >= r1 goto end
	Sub(cr, r4, r1)
	Movep(pc, ar, cr)
		Add(ans, r2, r3)	# ans = r2 + r3
 		Move(r3, r2)		# r3 = r2
 		Move(r2, ans)		# r2 = ans
 		Addi(r4, r4, 1)		# r4 += 1
		Subi(pc, pc, ?while)# return to while
#>end
#return ans
"""
```

```python
test = 20
print(fib_rec(test))
print(fib(test))

setr(r1, test)
start(fibasm)
print("ans: ",getr(ans))
```
