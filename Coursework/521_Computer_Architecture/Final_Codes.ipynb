```python
def mulpy(ina:r1, inb:r2)->r3:
    r3 = 0
    if inb >= 0:
        while inb!= 0:
            r3 += ina
            inb -= 1
        return r3
    else:
        while inb!= 0:
            r3 -= ina
            inb += 1
        return r3

multiASMCallby = f"""

#goto Main()
#:to_=Main
{go_to_}

#>mulasm(ina:r1, inb:r2)->r3:
#:x = r1
#:y = r2
#:z = ans

Load(x, st)
Subi(st, st, 1)

Load(y, st)
Subi(st, st, 1)

#:ra_ = x
#:rb_ = y
#:rd_ = z
{mul_rd_ra_rb_}

#save rd_ into the stack
Addi(st, st, 1)
Store(st, z)

#:to_ = returnhere
{go_to_}

#>Main():
    {pushrs_}
    #set inputs
    Set(r1, 7)
        Addi(st, st, 1)
        Store(st, r1)
    
    Set(r2, 5)
    Not(r2, r2)
    Addi(r2, r2, 1)
        Addi(st, st, 1)
        Store(st, r2)

    #:up_=mulasm
    {go_up_}

    #>returnhere
    Load(ans, st)
    Subi(st, st, 1)
    {poprs_}
    Move(ans, ans)
"""
start(multiASMCallby)
```

Fib recursive

```python
def fib(n):
    if n < 2:
        fn = n
    else:
        fn1 = fib(n-1)
        fn2 = fib(n-2)
        fn = fn1 + fn2
    return fn

fibrec = f"""
# jump to Main first
#:to_ = Main
{go_to_}

#> fib(n):
    # r5 used (for conditional 2)
    #: n = r1
    #: fn1 = r2
    #: fn2 = r3
    #: fn = r4
    #: retaddr = r9
    
    # get the return address - top most stack element
    Load(retaddr, st)
    Subi(st, st, 1)

    # get the n from stack - second top most stack element (receives the n from input, then from fn1 and fn2 recursively)
    Load(n, st)
    Subi(st, st, 1)
    
    # if n < 2:
    #: a_ = n
    #: b_ = r5
    #: end_ = else
    Set(r5, 2)
    {a_lt_b_}
        # fn = n
        Move(fn, n)
        #: to_ = endif
        {go_to_}

    #>else:
        # fn1 = fib(n-1)
        {pushrs_}       # store r0-r9 in stack
        # push n-1 into stack
        Subi(cr, n, 1)
        Addi(st, st, 1)
        Store(st, cr)
        # jump to fib
        #: fn_ = fib
        {call_fn_}
        # pop return value (value is returned if n < 2, otherwise keep recursion?)
        Load(res, st)
        Subi(st, st, 1)
        {poprs_}
        # put result into the var
        Move(fn1, res)

        # fn2 = fib(n-2)
        {pushrs_}
        # push n-2 into stack
        Subi(cr, n, 2)
        Addi(st, st, 1)
        Store(st, cr)
        # jump to fib
        #: fn_ = fib
        {call_fn_}
        # pop return value (value is returned if n < 2, otherwise keep recursion?)
        Load(res, st)
        Subi(st, st, 1)
        {poprs_}
        # put result into the var
        Move(fn2, res)

        # fn = fn1 + fn2
        Add(fn, fn1, fn2)

    #>endif
    # return fn
    # push fn into stack
    Addi(st, st, 1)
    Store(st, fn)
    # jump back to calling function
    Move(pc, retaddr) # where is the retaddr stored?

#> Main():
    # push input value for the function
    # Set(r6, 5)
    Addi(st, st, 1)
    Store(st, r6)
    # call
    #: fn_ = fib
    {call_fn_}
    # pop return value
    Load(r7, st)
    Subi(st, st, 1)
    # print
    Move(r7, r7)
"""

# testing
if __name__ == '__main__':
    al = []
    for n in range(10):
        Set(r6,n)
        start(fibrec)
        al = al + [getr(r7)]
    print(al)
    print([fib(n) for n in range(10)])

```
