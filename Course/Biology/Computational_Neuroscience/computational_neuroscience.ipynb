### 1.4 The electrical personality of neurons
The Idealized Neuron


Neuron: a leaky bag of charged liquid.
Has ion channels.
Neuron maintains potential difference across its membrane (more Na+, Cl-, H20 outside). Ionic pump expells Na+ out and K+ in.

Ionic channels are proteins that are selective and allow only specific ions to pass (Pass Na+ but not K+).
Ionic channels are Voltage-gated, Chemically-gated and Mechanically-gated.

If chemical from other neuron arrives at Synapse (junction between two neurons), chemically-gated channels open, Na+ from outside gets in neuron and changes local membrane potential. This causes opening/closing of voltage-gated channels in dendrites, body and axon, resulting in depolarization (+ change voltage) or hyperpolarization (- change in voltage - when K+ defuses outside).
Strong enough depolarization causes a spike or 'action potential'.

Action Potential: Na+ channels open > Na+ influx and more channel opens > spike > Na+ channels closes and K+ channel opens > K+ outflux restores membrane potential.
Action potential is propagated along the axon with Na+ open > Na+ close and K+ open > K+ closes.

Myelin allows fast long-range spike communication.


### 1.5 Making connections: synapses
Synapse: a connection/junction between two neurons.
- Electrical synapses use gap junctions: allows activity (e.g. Na+ action potential) of one neuron propagate directly in another neuron through ion channels. Found in places where simultaneous firing is needed.
- Chemical synapses use neurotransmitters: neurotransmitter molecule bag called vesicles. When action potential spike comes in along the axon of neuron A, the bags fuse with membrane and release neurotransmitter into the gap between two neurons > chemical gate ion channel opens > allows in/out of ions. This allows customized excitation in latter neuron.

A neuron can have up to 10,000 synapses. They can be:
- Excitatory: increases post-synaptic membrane potential. Input spike > neurotransmitter release (e.g., Glutamate) > bind to ion channel receptor > ion channels open > Na+ influx > depolarization due to EPSP (excitatory postsynaptic potential).
- Inhibitory:: decreases membrane potential of later neuron.

The Synapse Doctrine: synapses are the basis for memory and learning.

How do brains learn: synaptic plasticity.
Hebbian plasticity: if neuron A repeatedly takes part in firing neuron B, then the synapse from A to B is strengthened. It allows automatic/fast response upon stimuli.

Long Term Potentiation (LTP): increase in synaptic strength or EPSP size for same input over time.
Long Term Depression (LTD): decrease in synaptic strength or EPSP size for same input over time.



Time sensitivity is required for learnign sequences or making predictions.

### 2.3 Neural encoding: feature selection

P(response|stimulus) -> P(response|s1) as stimulus values can be overwhelming.

![Discretization](<attachment:10. Next.png>)

We'll take s(t) and discretize it: we'll segment the stimulus into n time points, plot values of each time point in an axis - so the dimension of the high dimension description equals n.
We want to sample the responses of the system to a variety of stimuli so we can characterize what it is abou the input that triggers responses. P(response|stimulus) -> P(response|s1, s2, ..., sn).
One common method is using Gaussian white noise - it choses a random number in each time step - covering all frequencies.

![Gaussian prior stimulus distribution](<attachment:11. Next.png>)

Each chunk of white noise of 100 time steps can be plotted in 100 dimensional space - the axis will be value at time t1, t2, etc. As we plot different samples of white noise, they're plotted at different points in the high-dimension space - and they start to fill a distribution that is itself multivariate Gaussian prior stimulus distribution. Along any axis, the distribution is Gaussian. Even if we take a new time point (taking a vector that is linear combination of existing dimensions), and project stimulus onto the new vector, the distribution will be Gaussian.

![Spike-conditional distribution](<attachment:12. Next.png>)

##### Nonlinear input/output function
The input/output function is: $P(spike|stimulus)->P(spike|s1)$
$P(spike|s_1)=\frac{P(s_1|spike)P(spike)}{P(s_1)}$

![P(spike|s1)](<attachment:16. Next.png>)

We can run some stimulus, collect some spikes, project the stimulus onto feature f1, extracting component s1. The histogram of s1 will be Gaussian (s1 covers normal distribution values) for white noise experiement. We then take values of s1 at times of spikes (skewed distribution). The spike inducing distribution is different from the prior. We take the ratio of P(s1|spike) or the later distribution by P(s1) or the prior distribution, and scale it by P(spike).

![Nonlinear input/output](<attachment:17. Next.png>)

If the neuron fires at random times (normally distributed), and compare it to normally distributed prior stimuli - we'll find both to have random normal distribution. This could mean the stimulus had nothing to do with firing of neuron, or we filtered out important attributes. We want to have high firing rate where projections in our identified feature are large.

##### Possibility of multiple features
There can be several filters, each selecting a different componenet of the input. It's possible that if those filters are present together, we get a large firing.

![Liniear/nonlinear model](<attachment:18. Next.png>)

![multiple filters](<attachment:19. Next.png>)

Finding multiple features: we took average of points and computed spike-triggered average. It's also possible to compute next order moment or its covariance. We apply Principle Component Analysis (PCA) for this, that finds low dimensional structure in high dimensional data. For example, cloud of data in 3D space can be 2D. PCA will give two principle components (for 2D) and orthogonal set of vectors that span the 2D cloud. Usually, the dimensions that pick out relavant structure in the data will be some linear combination of our stimulus coordinates in their original basis. PCA gives a new basis set (generally smaller than original) to represent our data.



Most faces can be represented with 7/8 PCAs called eigenfaces. If we have a new face that we want to fit, we can construct that as a linear combination of n eigenfaces. If we can express any face with some eigenfaces, we can express them using 7/8 numbers instead of hundreds of pixels.
PCA is also used to get sort out spike waves from 2/more neurons that were recorded on the same electrode. For two neurons, it can pick out two components that capture the largest amount of variance in the data. We then project each recording onto the two components. It'll keep the two neurons wave forms, and throw noise.

![PCA for spike sorting](<attachment:21. And throw noise.png>)

Each stimulus (blue dots) was 100 time steps of a white noise flicker applied on a retinal ganglion cell. PCA projected it in a 2D plain by projecting them onto the two found features. The spike trigger was close to zero - as there were two groups of stimuli that drove the neurons with average 0. But taking average of the groups gives us similar 'on' and 'off' features.

##### Finding good feature
- When the input/output curve over variable is interesting

Given filter f, when specific stimulus region given spike happens, the prior and the conditional distributions are distinct, and the input/output curve has interesting structure.

Can we find a filter f, so that when we project the stimulus onto it, the prior and conditional distributions are as different as possible.

The standard measure of difference between two probability distribution: DKL

$D_KL (P(S),Q(S))=\int dsP(s)log_2\frac{P(s)}{Q(s)}$

Goodness measure: maximise $D_KL (P(s_f |spike),P(s_f ))$
