## Linear transformation
A vector $\vec{v}=-1\hat{i}+2\hat{j}$ can be expressed as
$\left[\begin{array}{c}
-1\\
2
\end{array}\right]$ 

After a linear transformation, if $\hat{i}$ goes to $\hat{i}'$ and $\hat{j}$ goes to $\hat{j}'$, the vector $\vec{v}=-1\times \hat{i}'+2\times \hat{j}'$. 

The transformation is expressed with the matrix
$\left[\begin{array}{cc}
1 & 3\\
-2 & 0
\end{array}\right]$, indicating $\hat{i}$, which was initially
$\left[\begin{array}{c}
1\\
0
\end{array}\right]$, went to
$\left[\begin{array}{c}
1\\
-2
\end{array}\right]$. 
And $\hat{j}$, which was initially
$\left[\begin{array}{c}
0\\
1
\end{array}\right]$, went to
$\left[\begin{array}{c}
3\\
0
\end{array}\right]$.

A matrix indicating where $\hat{i}$ and $\hat{j}$ goes can be used to calculate wehre any vector in that space goes. 

So, $\vec{v}$ will end up at, 
$-1\times\left[\begin{array}{c}
1\\
-2
\end{array}\right]
+2\times\left[\begin{array}{c}
3\\
0
\end{array}\right]=
\left[\begin{array}{c}
-1(1)+2(3)\\
-1(-2)+2(0)
\end{array}\right]=
\left[\begin{array}{c}
5\\
2
\end{array}\right]$

### 1.1 The Geometry of Linear Equations
The fundamental problem of linear algebra: solve a system of linear equations, i.e., solve $n$ linear equations in $n$ unknows. For example:
$2x-y=0$  
$-x+2y=3$

This problem can be viewed in three ways.

#### Row pictures


```python
import matplotlib.pyplot as plt
import numpy as np

def eq1(x):
    return 2*x

def eq2(x):
    return (3+x)/2

# Solution
x_intersect = np.linalg.solve([
    [2, -1],
    [-1, 2]
], [0, 3])

y_intersect = eq1(x_intersect[0])

# Generate x values
x = np.linspace(-2, 5, 100)

# Calculate y values for each equation
y1 = eq1(x)
y2 = eq2(x)

# Plot the lines
plt.plot(x, y1, label="2x-y=0")
plt.plot(x, y2, label="-x+2y=3")

# Solution
print("Intersection point:", (x_intersect[0], y_intersect))

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

For 3 linear equations with 3 unknowns, three planes will intersect in a point in row picture.

#### Matrix Multiplication

$Ax=b$

$\left[\begin{array}{cc}

This technique shows that $A\mathbf{x}$ is a linear combination of the columns of $A$.

We can also calculate $A\mathbf{x}$ by taking the dot product of each row of $A$ with the vector $\mathbf{x}$.

#### Linear Independence
In the equation  
$A\mathbf{x}=b$  
If we can solve every possible vector $\mathbf{b}$ by taking combinations of $A$, or if all combinations of the columns of $A$ fill the entire 2-d plane, then the columns of $A$ are independent.  
Otherwise, $A$ is a *singular matrix*, and the columns of $A$ are *linearly dependent*.

#### Subspaces

Assuming the last column of the matrix $C$ to be $w^*$, geometrically, $w^*$ is in the same plane of $u, v$. Thus, the 3 columns are not independent and don't span the full 3-d space (they are dependent, while the columns of $A$ are independent).

There are many vectors in $\mathbf{R}^3$ which do not lie in that plane. Those vectors can't be written as a linear combination of the columns of $C$, and so correspond to values of b for which $C\mathbf{x}=b$ has no solution $\mathbf{x}$. The linear combinations of the columns of $C$ form a two dimensional *subspace* of $\mathbf{R}^3$.

But the all linear combinations of the colunmns of matrix $A$,
$\mathbf{u}=
\left[\begin{array}{c}
1\\
-1\\

\end{array}
\right],
\mathbf{v}=
\left[\begin{array}{c}

\end{array}
\right],
\mathbf{w}=
\left[\begin{array}{c}
0\\
0\\
1
\end{array}
\right]$

span the entire space $\mathbf{R}^3$; the equation $A\mathbf{x}=b$ has solution for every $\mathbf{b}$ in $\mathbf{R}^3$. We say that $\mathbf{u}, \mathbf{v}, \mathbf{w}$ form a *basis* for $\mathbf{R}^3$.

A basis for $\mathbf{R}^n$ is
- a collection of $n$ independent vectors in $\mathbf{R}^n$, or
- a collection of $n$ vectors whose combinations cover teh whole $n$-dimensional space, or
- a collection of basis vectors when a matrix which has those vectors as its columns is invertible.

A *vector space* is a collection of vectors that is closed under linear combinations. A *subspace* is a vector space inside another vector space. In a 3-d space, the subspaces are:
- whole 3-d space,
- 2-d plane through the origin,
- 1-d line through the origin,
- origin point.

Matrices can be rectangular; we can have seven equations in three unknowns. Rectangular matrices are not invertible, but the symmetric, square matrix $A^{T}A$ may be invertible.

### 1.3 Elimination with Matrices
#### Gaussian Elimination:
*Elimination* is the technique most commonly used by computer software to solve systems of linear equations. It finds a solution $\mathbf{x}$ to $A\mathbf{x}=\mathbf{b}$ whenever the matrix $A$ is invertible. For example, for the linear equation:

$x+2y+z=2$  
$3x+8y+z=12$  
$4y+z=2$

$A\mathbf{x}=b$

$\left[\begin{array}{ccc}
1 & 2 & 1\\
3 & 8 & 1\\
0 & 4 & 1
\end{array}\right]
\left[\begin{array}{c}
x\\
y\\
z
\end{array}\right]=
\left[\begin{array}{c}
2\\
12\\
2
\end{array}\right]$

$A_{11}$ is the first pivot, $A_{22}$ is the second pivot, $A_{33}$ is the third pivot. Pivots are never $0$. If there is a $0$ in the pivot position, we'll try to fix it by switching rows. *Temporary failure* is when we can perform row exchange to get non-zero value in pivot position. If we can't do that, it's *complete failure*.

##### Forward elimination

$\left[\begin{array}{ccc|c}

\end{array}\right]
=>(2,1): (2)-3(1)=>
\left[\begin{array}{ccc|c}

0 & 4 & 1 & 2
\end{array}\right]
=>(3,2): (3)-2(2)=>
\left[\begin{array}{ccc|c}

\end{array}\right]$

Here,
$\left[\begin{array}{ccc}

$\left[\begin{array}{c}

The *determinant* of $u$ is the product of the pivots.

The equations we get:

$5z=-10$

##### Back substitution

$5z=-10=>z=-2$

### 1.5 Factorization into $A=L\mathcal{U}$

#### Inverse of a product

$(AB)^{-1}=B^{-1}A^{-1}$

Thus,

$AB(AB)^{-1}=ABB^{-1}A^{-1}=AIA^{-1}=AA^{-1}=I$

or,

$(AB)^{-1}AB=B^{-1}A^{-1}AB=BIB^{-1}=BB^{-1}=I$

#### Transpose of a product
We obtain the *transpose* of a matrix by exchanging its rows and columns. The entry is row *i* column *j* of *A* is the entry in row *j* column *i* of $A^T$.
The transpose of a matrix product $AB$ is $B^TA^T$. For any invertible matrix $A$, the inverse of $A^T$ is $(A^{-1})^T$.

#### $A=L\mathcal{U}$

#### Complexity of elimination operation
While using elimination to find $A=L\mathcal{U}$, we nned to keep track of row multiplication and subtractions. We need to remember $L$ that becomes $\mathcal{U}$, we don't have to store $A$ or $E_{ij}$ in computer's memory.


For a $n\times n$ matrix, eliminating a value requires scaling one row and subtracting it from another, which requires on the order of $n$ operations. Eliminating all values in the first column (n rows minus the first row), we need to multiply+subtract over all rows. So we require about $n^2$ operations. After finishing column 1, we'll perform $(n-1)^2$ operations for the shorter column 2.

Cost of $A$:  
$n^2+(n-1)^2+(n-2)^2+...+1^2={n^3\over3}$

Cost of $b$ is $n^2$.
