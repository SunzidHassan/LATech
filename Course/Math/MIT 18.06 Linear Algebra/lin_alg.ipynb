### 1.1 The Geometry of Linear Equations
The fundamental problem of linear algebra: solve a system of linear equations.

#### Row pictures
$2x-y=0$  
$-x+2y=3$

```python
import matplotlib.pyplot as plt
import numpy as np

def eq1(x):
    return 2*x

def eq2(x):
    return (3+x)/2

# Solution
x_intersect = np.linalg.solve([
    [2, -1],
    [-1, 2]
], [0, 3])

y_intersect = eq1(x_intersect[0])

# Generate x values
x = np.linspace(-2, 5, 100)

# Calculate y values for each equation
y1 = eq1(x)
y2 = eq2(x)

# Plot the lines
plt.plot(x, y1, label="2x-y=0")
plt.plot(x, y2, label="-x+2y=3")

# Solution
print("Intersection point:", (x_intersect[0], y_intersect))

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

For 3 linear equations with 3 unknowns, three planes will intersect in a point in row picture.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the vector
c1 = np.array([2, -1])
c2 = np.array([-1, 2])
s = np.array([0, 3])

# Plot the vector with arrow
plt.figure()
plt.arrow(0, 0, c1[0], c1[1], head_width=0.1, head_length=0.1, fc='green', ec='green')
plt.arrow(0, 0, c2[0], c2[1], head_width=0.1, head_length=0.1, fc='blue', ec='blue')
plt.arrow(2, -1, c2[0], c2[1], head_width=0.1, head_length=0.1, fc='blue', ec='blue')
plt.arrow(1, 1, c2[0], c2[1], head_width=0.1, head_length=0.1, fc='blue', ec='blue')
plt.arrow(0, 0, s[0], s[1], head_width=0.1, head_length=0.1, fc='red', ec='red')

plt.xlabel('x')
plt.ylabel('y')

# Make x and y axis bold
plt.axhline(0, color='gray', linewidth=1)
plt.axvline(0, color='gray', linewidth=1)

plt.grid()
plt.show()
```
