Receptive field: specific properties of a sensory stimulus that generate a strong response from the cell.

The light data is processed first in 'Lateral Geniculate Nucleus (LGN)', and finally in 'Primary Visual Cortex (V1)'.

Descriptive model:
in LGN there are:
LGN RF - On-center, off-surround receptive fields: neuron spikes if light is on in the center of a retina cell field, and off in the surrounding.
LGN RF - Off-center, on-surround receptive fields: neuron spikes if light is off in the center of a retina cell field, and on in the surrounding.

V1 RF: there are many oriented (45', horizontal, vertical) receptive on/off center receptive fields.

How did oriented receptive fields obtained from center-surround receptive fields?

Mechanistic model: V1 RFs are created from converging LGN inputs (feed forward) and recurrent inputs.



Why are RFs in V1 shpaed in this way?
Interpretive model:
Efficient coding hypothesis: represent images as faithfully and efficiently as possible using neurons with receptive fields RF1, RF2, etc.
Given image I, we can reconstruct I-hat using neural responses r1, r2,...  

$\hat{I}=\sum \limits _{i}RF_{i}r_{i}$

What are the $RF_{i}$ that minimize total squared pixelwise errors between $I$ and $\hat{I}$ and are as independent as possible?
